```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Number of pet':[12,23,8,17,14,7,90,5,13,56],
        'Person':['Person 1','Person 2','Person 3','Person 4','Person 5','Person 6','Person 7','Person 8','Person 9','Person 10']}#csv file import korte hobe
df = pd.DataFrame(data)
df
```

```python
#data sorting korte hbe 
df = df.sort_values(by='Number of pet',ascending=False)
df
```

```python
#data slice korte hobe top 5 data dkhte 
df=df.head(5)
df
```

```python
#color thik kore dite hobe jeta ami bar e use korte chai
colors = ['#3A1078','#4E31AA','#2F58CD','#3795BD','#191825']
```

```python
#horizontal bar plot create korte hobe #barh horizontal bar indicate kore karon eta horizontal bar create kore 
fig,ax=plt.subplots()
bars = ax.barh(df['Person'],df['Number of pet'],color=colors)

#plot customize korte hobe 
#enumerate for loop er maddhome 
for i,bar in enumerate(bars):
  ax.text(bar.get_width() + 0.10,bar.get_y()+bar.get_height()/2,df['Number of pet'].iloc[i],ha='center',va='center',fontsize=11)
ax.set_xlabel('Number of pet')
ax.set_title('Number of pet owned by each person')

#plot dekhaite hobe
plt.show()
```
